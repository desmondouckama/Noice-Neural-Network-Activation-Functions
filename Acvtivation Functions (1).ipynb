# Activation Functions

## Step Function

$\varphi(v)=\left\{\begin{array}{l}1, v \geq 0 \\ 0, v<0\end{array} \Rightarrow y_k=\left\{\begin{array}{ll}1, & v_k \geq 0, u_k \geq \theta_k \\ 0, & v_k<0, u_k<\theta_k\end{array}\right\}\right.$

### Without Using "For Loops:"

```python
import matplotlib.pyplot as plt
import numpy as np

#defines the STEP FUNCTION
def stepFunction(neuron_input_value):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value >= 0)] = 1
    neuron_output_value[np.where(neuron_input_value < 0)] = 0
    return neuron_output_value

#plots the STEP FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,100)
plt.plot(neuron_input_value, stepFunction(neuron_input_value))
plt.title("Step Function")

#Sends to a PDF for reference 
plt.savefig("Step_Fun_Graph.pdf")

        
        
```

## Step Linear Functions
$$
    \varphi(v)=\left\{\begin{array}{l}
    1, v \geq a \\
    v,-a<v<a \\
    0, v<-a
    \end{array}\right\}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

#defines the STEP LINEAR FUNCTION
def stepLinearFunction(neuron_input_value, step_size):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value >= step_size)] = 1
    neuron_output_value[np.where(neuron_input_value < step_size)] = neuron_input_value[np.where(neuron_input_value < step_size)]
    neuron_output_value[np.where(-1*step_size < step_size)] = neuron_input_value[np.where(-1*step_size < step_size)] 
    neuron_output_value[np.where(neuron_input_value < -1*step_size)] = 0
    return neuron_output_value

#plots the STEP LINEAR FUNCTION over a range of NEURON_INPUT_VALUES for different values of a
neuron_input_value = np.linspace(-2,2,32)
fig, axs = plt.subplots(2,2)
fig.tight_layout(pad=3.0)
def plot1(ax, data):
    ax.plot(data, stepLinearFunction(data, 0))
    ax.set_title("Step Linear Function (a = 0)")
def plot2(ax, data):
    ax.plot(data, stepLinearFunction(data, 1))
    ax.set_title("Step Linear Function (a = 1)")
def plot3(ax, data):
    ax.plot(data, stepLinearFunction(data, 8))
    ax.set_title("Step Linear Function (a = step-size)")
    
plot1(axs[0, 0], neuron_input_value)
plot2(axs[0, 1], neuron_input_value)
plot3(axs[1, 0], neuron_input_value)
axs[1, 1].axis('off')




#Sends to a PDF for reference 
plt.savefig("Step_Linear_Graph.pdf")
```
