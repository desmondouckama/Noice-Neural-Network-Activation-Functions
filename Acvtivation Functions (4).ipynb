# Activation Functions

## Step Function

$$\varphi(v)=\left\{\begin{array}{l}1, v \geq 0 \\ 0, v<0\end{array} \Rightarrow y_k=\left\{\begin{array}{ll}1, & v_k \geq 0, u_k \geq \theta_k \\ 0, & v_k<0, u_k<\theta_k\end{array}\right\}\right.$$

### Without Using "For Loops:"

```python
import matplotlib.pyplot as plt
import numpy as np

#defines the STEP FUNCTION
def stepFunction(neuron_input_value):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value >= 0)] = 1
    neuron_output_value[np.where(neuron_input_value < 0)] = 0
    return neuron_output_value

#plots the STEP FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,100)
plt.plot(neuron_input_value, stepFunction(neuron_input_value))
plt.title("Step Function")

#Sends to a PDF for reference 
plt.savefig("Step_Fun_Graph.pdf")

        
        
```

## Step Linear Functions
$$
    \varphi(v)=\left\{\begin{array}{l}
    1, v \geq a \\
    v,-a<v<a \\
    0, v<-a
    \end{array}\right\}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

#defines the STEP LINEAR FUNCTION
def stepLinearFunction(neuron_input_value, step_size):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value >= step_size)] = 1
    neuron_output_value[np.where(neuron_input_value < step_size)] = neuron_input_value[np.where(neuron_input_value < step_size)]
    neuron_output_value[np.where(-1*step_size < step_size)] = neuron_input_value[np.where(-1*step_size < step_size)] 
    neuron_output_value[np.where(neuron_input_value < -1*step_size)] = 0
    return neuron_output_value

#plots the STEP LINEAR FUNCTION over a range of NEURON_INPUT_VALUES for different values of a
neuron_input_value = np.linspace(-2,2,32)
fig, axs = plt.subplots(2,2)
fig.tight_layout(pad=3.0)
def plot1(ax, data):
    ax.plot(data, stepLinearFunction(data, 0))
    ax.set_title("Step Linear Function (a = 0)")
def plot2(ax, data):
    ax.plot(data, stepLinearFunction(data, 1))
    ax.set_title("Step Linear Function (a = 1)")
def plot3(ax, data):
    ax.plot(data, stepLinearFunction(data, 8))
    ax.set_title("Step Linear Function (a = step-size)")
    
plot1(axs[0, 0], neuron_input_value)
plot2(axs[0, 1], neuron_input_value)
plot3(axs[1, 0], neuron_input_value)
axs[1, 1].axis('off')

#Sends to a PDF for reference 
plt.savefig("Step_Linear_Graph.pdf")
```

## Sigmoidal Functions
$$
\varphi(v)=\frac{1}{1+e^{-a v}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt

#defines the SIGMOIDAL
def sigmoid(neuron_input_value, a):
    neuron_output_value = np.copy(neuron_input_value)
    A = -1*a*neuron_input_value
    B = mt.e**A
    C = 1/(1+B)
    return C

#plots the STEP FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,64)
fig, axs = plt.subplots(2,2)
fig.tight_layout(pad=3.0)
def plot1(ax, data):
    ax.plot(data, sigmoid(data, 0))
    ax.set_title("Step Linear Function (a = 0)")
def plot2(ax, data):
    ax.plot(data, sigmoid(data, 4))
    ax.set_title("Step Linear Function (a = 4)")
def plot3(ax, data):
    ax.plot(data, sigmoid(data, 16))
    ax.set_title("Step Linear Function (a = step-size)")
    
plot1(axs[0, 0], neuron_input_value)
plot2(axs[0, 1], neuron_input_value)
plot3(axs[1, 0], neuron_input_value)
axs[1, 1].axis('off')

#Sends to a PDF for reference 
plt.savefig("Sigmoidal.pdf")
```

## Rectified Linear Unit (ReLU)
$$
\operatorname{relu}(x)=\max (0, x)
$$

```python
import matplotlib.pyplot as plt
import numpy as np

#defines the RELU FUNCTION
def reluFunction(neuron_input_value):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value > 0)] = neuron_output_value[np.where(neuron_input_value > 0)]
    neuron_output_value[np.where(neuron_input_value <= 0)] = 0
    return neuron_output_value

#plots the ReLU FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,100)
plt.plot(neuron_input_value, reluFunction(neuron_input_value))
plt.title("RELU Function")

#Sends to a PDF for reference 
plt.savefig("RELU_Graph.pdf")

        
        
```

## Softplus Function
$$
\operatorname{Softplus}(x)=\frac{1}{\beta} \cdot \log (1+\exp (\beta \cdot x))
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt

#defines the softplus
def softplus(neuron_input_value, a):
    neuron_output_value = np.copy(neuron_input_value)
    A = 1 + mt.e**(a*neuron_output_value)
    B = np.log(A)/a
    return B

#plots the softplus FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,64)
fig, axs = plt.subplots(2,2)
fig.tight_layout(pad=3.0)
def plot1(ax, data):
    ax.plot(data, softplus(data, 1))
    ax.set_title("Softplus Function (a = 1)")
def plot2(ax, data):
    ax.plot(data, softplus(data, 4))
    ax.set_title("Softplus Function (a = 4)")
def plot3(ax, data):
    ax.plot(data, softplus(data, 16))
    ax.set_title("Softplus Function (a = step-size)")
    
plot1(axs[0, 0], neuron_input_value)
plot2(axs[0, 1], neuron_input_value)
plot3(axs[1, 0], neuron_input_value)
axs[1, 1].axis('off')

#Sends to a PDF for reference 
plt.savefig("softplus.pdf")
```

## RELU and Softplus

```python
import matplotlib.pyplot as plt
import numpy as np

#defines both
def reluFunction(neuron_input_value):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value > 0)] = neuron_output_value[np.where(neuron_input_value > 0)]
    neuron_output_value[np.where(neuron_input_value <= 0)] = 0
    return neuron_output_value
def softplus(neuron_input_value, a):
    neuron_output_value = np.copy(neuron_input_value)
    A = 1 + mt.e**(a*neuron_output_value)
    B = np.log(A)/a
    return B

#plots both
neuron_input_value = np.linspace(-1,1,100)
plt.plot(neuron_input_value, reluFunction(neuron_input_value), label = "RELU")
plt.plot(neuron_input_value, softplus(neuron_input_value, 4), label = "Softplus")
plt.legend(loc="upper left")

plt.title("RELU Function and Softplus")

#Sends to a PDF for reference 
plt.savefig("RELU_Softplus.pdf")
```
