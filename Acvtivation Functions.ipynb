# Activation Functions

## Step Function

$\varphi(v)=\left\{\begin{array}{l}1, v \geq 0 \\ 0, v<0\end{array} \Rightarrow y_k=\left\{\begin{array}{ll}1, & v_k \geq 0, u_k \geq \theta_k \\ 0, & v_k<0, u_k<\theta_k\end{array}\right\}\right.$

### Without Using "For Loops:"

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#defines the STEP FUNCTION
def stepFunction(neuron_input_value):
    neuron_output_value = np.copy(neuron_input_value)
    neuron_output_value[np.where(neuron_input_value >= 0)] = 1
    neuron_output_value[np.where(neuron_input_value < 0)] = 0
    return neuron_output_value

#plots the STEP FUNCTION over a range of NEURON_INPUT_VALUES
neuron_input_value = np.linspace(-2,2,100)
plt.plot(neuron_input_value, stepFunction(neuron_input_value))
plt.title("Step Function")

#Sends to a PDF for reference 
plt.savefig("Step_Fun_Graph.pdf")

        
        
```
